```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ptitprince as pt

data = pd.read_csv('pima-diabetes (1).csv')
df = pd.DataFrame(data)

outCome0 = df[df["Outcome"] == 0]
outCome1 = df[df["Outcome"] == 1]

# BOX PLOT FOR AGE COMPARING OUTCOME 1 WITH 0
plt.boxplot([outCome0["Age"], outCome1["Age"]], vert=1, labels=['Outcome 0', 'Outcome 1'])
plt.title("Boxplot for Age")
plt.show()

# CONFIRM WITH VIOLIN PLOTS
plt.violinplot([outCome0["Age"], outCome1["Age"]], showmedians=True)
plt.xticks([1, 2], ['Outcome 0', 'Outcome 1'])
plt.ylabel("Age")
plt.title("Violinplot for Age")
plt.show()

#PLOT RAINCLOUD
plt.title("Raincloud Plot for Age")
plt.xlabel("Age")
pt.RainCloud(data=[outCome0["Age"], outCome1["Age"]], bw=0.05, cut=0, orient='h')
plt.show()

# BOX PLOT FOR GLUCOSE COMPARING OUTCOME 1 WITH 0
plt.boxplot([outCome0["Glucose"], outCome1["Glucose"]], vert=1, labels=['Outcome 0', 'Outcome 1'])
plt.title("Boxplot for Glucose")
plt.show()

# CONFIRM WITH VIOLIN PLOTS
plt.violinplot([outCome0["Glucose"], outCome1["Glucose"]], showmedians=True)
plt.xticks([1, 2], ['Outcome 0', 'Outcome 1'])
plt.ylabel("Glucose")
plt.title("Violinplot for Glucose")
plt.show()
#PLOT RAINCLOUD
plt.title("Raincloud Plot for Glucose")
plt.xlabel("Age")
pt.RainCloud(data=[outCome0["Glucose"], outCome1["Glucose"]], bw=0.05, cut=0, orient='h')
plt.show()
```
