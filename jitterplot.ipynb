```python
#question 7
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('median-household-income (1).csv', encoding = 'utf-16', sep='\t')
df = pd.DataFrame(data)
#references https://seaborn.pydata.org/generated/seaborn.stripplot.html
plt.figure(figsize=(12, 6))
sns.stripplot(data =df, x='State',y='Estimate',palette="deep")
plt.gca().tick_params(axis='x', which='major')
plt.title("Visualizing Median Income For All 3000+ US Counties")
plt.xlabel("State", labelpad=100)
plt.ylabel("Mdeian Income in $")
plt.xticks(rotation=90)
print(df)
```
