```python
#Author: Rohan Keenoy
#CS 3411
#Some code references this source: https://python-graph-gallery.com/390-basic-radar-chart/
#such as list flattening and angles calculations and initializing the plot
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# i was just estimating off the graph for the data, it had different ticks for each column so for simplification purposes
# I kept it all in a range of 0-100%
playerData = {
    'Player': ['Player 1', 'Player 2', 'Player 3'],
    'Turnovers': [100, 30, 80],
    'Fouls': [60, 100, 30],
    'Assists': [10, 50, 100],
    'Rebounds': [10, 50, 100],
    'Assist Turnover': [20, 100, 20],
    'Steals': [50, 100, 10]
}

df = pd.DataFrame(playerData)

# need to get the number of categories each player has
categories = list(df.columns[1:])
N = len(categories)

# Calculate angle for each axis
angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]

#initialize the radar chart
ax = plt.subplot(111, polar=True)

#iterate the the player categories onto the graph
for i, row in df.iterrows():
    values = row.drop('Player').values.flatten().tolist()
    values += values[:1]

    plt.plot(angles, values, linewidth=1, linestyle='solid', label=row['Player'])
    plt.fill(angles, values, alpha=0.1)

# Add legend and anchor to the top right
plt.legend(bbox_to_anchor=(1.3, 1.0))
# Customize the plot
plt.xticks(angles[:-1], categories, color='grey', size=12)
#had to add padding because the words were in the chart itsself
plt.tick_params(axis='x', pad=20)
plt.ylim(0, 100)

plt.title('Player Comparison Radar Chart', pad=15,size=20)

# Show the radar chart
plt.show()

```
