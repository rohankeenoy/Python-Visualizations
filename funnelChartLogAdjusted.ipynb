```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("county_demographics.csv")
df = pd.DataFrame(data)

plt.scatter(x=df["Population.2020 Population"], y=df["Education.Bachelor's Degree or Higher"])
plt.title("Funnel plot of bachelors degrees in 2020 population with log scale")
plt.ylabel("Percent with bachelors")
plt.xlabel("Population of County")

#Set y-axis to use a log scale so that the data was more easily accessible.
plt.xscale('log')
plt.show()


plt.scatter(x=df["Population.2020 Population"], y=df["Education.Bachelor's Degree or Higher"])
plt.title("Funnel plot of bachelors degrees in 2020 population without log scale")
plt.ylabel("Percent with bachelors")
plt.xlabel("Population of County")

plt.ticklabel_format(style='plain', axis='x')




plt.show()
```

Without scaling the population, the plot showed that lower population areas had a large spread in their data due to a small population showing more variation.The larger the population the less variation there was so larger county populations seemed to appear as lower, when accounting for adjusted graduation rates, it appears the percent with bachelors is much more closer together because there is not as much variance from the sample.
