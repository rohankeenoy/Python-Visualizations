```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("county_demographics.csv")
df = pd.DataFrame(data)
df.head()

dfMo = df[df['State'] == 'MO']
dfWa = df[df['State'] == 'WA']
dfWi = df[df['State'] == 'WI']
dfMA = df[df['State'] == 'MA']
dfCO = df[df['State'] == 'CO']

dfs = [dfMo, dfWa, dfWi, dfMA, dfCO]
allStates = pd.concat(dfs)
print(allStates.head())

plt.boxplot([group['Population.2020 Population'] for _, group in allStates.groupby('State')],
            labels=allStates['State'].unique())
plt.title("Non Standardized 5 State Population Distribution")
plt.xlabel('State')
plt.ylabel('Standardized Population')
plt.show()

plt.boxplot([group['Population.2020 Population'] for _, group in allStates.groupby('State')],
            labels=allStates['State'].unique())
plt.title("Standardized 5 State Population Distribution")
plt.xlabel('State')
plt.ylabel('Standardized Population')

plt.yscale('log')

plt.show()
```
