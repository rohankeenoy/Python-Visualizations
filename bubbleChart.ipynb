```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib.ticker import ScalarFormatter
data = pd.read_csv("traffic-fatality.csv")
df = pd.DataFrame(data)

min_size = df['Total fatality last year'].min()
max_size = df['Total fatality last year'].max()
sizes = np.linspace(20, 200, len(df))

plt.figure(figsize=(12, 8))
df_sorted = df.sort_values(by='Total fatality last year', ascending=False)
scatterplot = sns.scatterplot(data=data, x="Fatality per 100K inhabitants per year", y="Fatailty per 100K motor vehicles per year", hue="Total fatality last year", size=sizes, alpha=0.7)

plt.title("Traffic Fatality Bubble Chart")
plt.xlabel("Fatality per 100K inhabitants per year")
plt.ylabel("Fatality per 100K motor vehicles per year")

# Get the unique country names
countries = df_sorted['North American Country'].unique()

# Create a legend with country names
legend_labels = []
for country in countries:
    legend_labels.append(f'{country}: {df_sorted[df_sorted["North American Country"] == country]["Total fatality last year"].values[0]}')

# Set the legend with country names
plt.legend(legend_labels,title="Countries", bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)

# Set the x and y scales to log

plt.yscale("log")
formatter = ScalarFormatter()
formatter.set_scientific(False)
formatter.set_useOffset(False)
plt.gca().yaxis.set_major_formatter(formatter)
# Show the plot
plt.show()
```
