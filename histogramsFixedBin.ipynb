```python
#number5
import pandas as pd
import matplotlib.pyplot as plt
import math

data = pd.read_csv("pima-diabetes.csv")
df = pd.DataFrame(data)
df.head()
age = df.Age

#calculating number of bins and bin width:
#references https://www.qimacros.com/histogram-excel/how-to-determine-histogram-bin-interval/
n = len(age)
numBins = math.ceil(math.sqrt(n))
min = age.min()
max = age.max()

plt.hist(age, bins=numBins, color="lightseagreen", edgecolor="k", range=(min,max))
plt.xlabel("Age")
plt.ylabel("Count of Age")
plt.title(f"Histogram of Age (bin size = {numBins})")

plt.grid(alpha=0.7)
#sets the axis below the bars
plt.gca().set_axisbelow(True)
plt.show()
```

```python
#number6
bmi = df.BMI
print(bmi)
#calculating number of bins and bin width:
#references https://www.qimacros.com/histogram-excel/how-to-determine-histogram-bin-interval/
n = len(bmi)
numBins = math.ceil(math.sqrt(n))
min = bmi.min()
max = bmi.max()

plt.hist(age, bins=numBins, color="lightgreen", edgecolor="k", range=(min,max))
plt.xlabel("BMI (bin)")
plt.ylabel("Count of BMI")
plt.title(f"BMI Histogram(bin size = {numBins})")
plt.grid(axis='y',alpha=0.7)
#sets the axis below the bars
plt.gca().set_axisbelow(True)
plt.show()
```
