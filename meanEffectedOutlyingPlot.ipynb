```python
import matplotlib.pyplot as plt
import pandas as pd

data = [100, 85, 95, 105, 110, 700]
df = pd.DataFrame(data)
mean = df.mean()[0]  # Calculate the mean of the DataFrame

plt.plot(df.index, df[0], label='Data')
plt.axhline(mean, color='r', linestyle='--', label='Mean')
plt.legend()
plt.title("How mean may not be completely honest in average")
plt.show()

median = df.median()[0]
print(median)

plt.plot(df.index, df[0], label='Data')
plt.axhline(median, color='r', linestyle='--', label='median')
plt.legend()
plt.title("The median is much more fair when there are outliers within data")
plt.show()

print("The median's assessment on this dataset with a significant outlier is much better judge of how this theoretical income may be precieved\n By using mean it is under the influence of large outliers and may not be the rest way to judge tax rates for example.  ")
```
