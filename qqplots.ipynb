```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

data = pd.read_csv('pima-diabetes (1).csv')
df = pd.DataFrame(data)

age = df.Age
bp = df.BloodPressure
sortAge = np.sort(age)
sortB = np.sort(bp)

sm.qqplot(sortAge, line='s')
plt.title('Q-Q Plot for Age')
plt.show()


sm.qqplot(sortB, line='s')
plt.title('Q-Q Plot for BloodPressure')
plt.show()
```

Because blood pressure has so many values for 0, it can be said that age is more normal distributed, however, both plots are not perfect. The Blood Pressure one is heavily Right scewed while the age qqplot appears to be underdispersed.
